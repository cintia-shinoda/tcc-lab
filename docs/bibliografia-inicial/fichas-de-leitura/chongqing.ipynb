| Referência ABNT | SONG, J.; DING, J.; GUI, X.; ZHU, Y. Assessment and solutions for vulnerability of urban rail transit network based on complex network theory: A case study of Chongqing. Heliyon, v. 10, n. 5, e27237, 2024. DOI: 10.1016/j.heliyon.2024.e27237 | 
| Autores | Jinghua Song, Jianfeng Ding (Wuhan University), Xuechen Gui (Wuhan University), Yuyi Zhu (Chongqing University) |
| Ano / Periódico | 2024 / Heliyon (Elsevier) — open access, peer-reviewed |
| Objetivo | Avaliar vulnerabilidades e pontos fracos da rede de metrô de Chongqing usando teoria de redes complexas, classificar nós por nível de criticidade e propor estratégias de gestão diferenciadas |
| Metodologia | Modelo topológico construído com método L-space. Indicadores calculados: degree, betweenness centrality, eficiência de rede e volume de fluxo de passageiros (PFV). Classificação dos nós em 6 clusters usando K-means. Simulação de ataque: remoção individual de cada nó e medição da redução na eficiência global da rede. Análise da distribuição geográfica dos nós críticos considerando a topografia montanhosa de Chongqing. Rede: 233 nós, 254 arestas, grau médio 2,197 |
| Principais resultados | (1) 6 clusters de nós com perfis de vulnerabilidade distintos. (2) A remoção dos nós mais críticos causa redução de mais de 6% na eficiência global. (3) Achado-chave: a importância de um nó NÃO corresponde necessariamente ao seu ranking de degree ou betweenness. Exemplo: Chongqing North Station tem degree 4 e betweenness 3º, mas sua remoção reduz eficiência em apenas 3%. Cuntan tem degree 2 e betweenness 24º, mas sua remoção causa redução de 8%. (4) 82,85% dos nós têm degree 2 (rede linear). (5) Propostas de estratégias de gestão diferenciadas por cluster |
| Relação com o TCC | Conexão direta e forte. (1) Usa exatamente a representação L-space e as mesmas métricas que nós (degree, betweenness, eficiência). (2) O achado de que degree/betweenness não predizem perfeitamente a vulnerabilidade real é crucial — confirma o que encontramos em SP (degree ≠ betweenness). (3) O uso de K-means pra classificar nós é uma possibilidade de extensão do nosso trabalho. (4) Grau médio de 2,197 é similar ao nosso 2,67 — redes predominantemente lineares |
| Citações úteis | Os autores demonstram que a importância de um nó para a eficiência global da rede nem sempre corresponde ao seu ranking em métricas tradicionais de centralidade, evidenciando a necessidade de avaliações multidimensionais |
| Limitações | (1) Foca em metrô, não ônibus. (2) Rede pequena (233 nós vs nossos 22.103). (3) Não avalia equidade territorial. (4) Dados de PFV restritos a uma semana |
| Usar no TCC como | Referência metodológica — justifica o uso de múltiplas métricas e a classificação de nós. O contraste degree vs impacto real é um argumento forte pro nosso TCC. Citar na introdução e na discussão dos resultados |
