# Topological assessment of recoverability in public transport networks


| Campo | Conteúdo |
|---|---|
| Referência ABNT | MASSOBRIO, R.; CATS, O. Topological assessment of recoverability in public transport networks. Communications Physics, v. 7, art. 108, 2024. DOI: 10.1038/s42005-024-01596-8 |
| Autores | Renzo Massobrio (TU Delft), Oded Cats (TU Delft) |
| Ano / Periódico | 2024 / Communications Physics (Nature) — alto impacto |
| Objetivo | Introduzir uma abordagem topológica para avaliar a recoverability (capacidade de recuperação) de redes de transporte público após perturbações, e relacionar indicadores topológicos com propriedades de recuperação |
| Metodologia | Análise de 42 redes de metrô ao redor do mundo. Grafos construídos incorporando características de infraestrutura e serviço (tempo de viagem, tempo de espera, transferências). Perturbações modeladas como remoção progressiva de links (25%, 50%, 75%, 100%). Indicadores topológicos avaliados: número de nós, número de links, diâmetro, densidade, meshedness, caminho médio mais curto (ponderado e não ponderado), distribuição de betweenness centrality (taxa da distribuição exponencial), assortatividade |
| Principais resultados | (1) Alta correlação entre indicadores topológicos e capacidade de recuperação. (2) Redes mais densas e eficientes (menor caminho médio) resistem melhor a perturbações. (3) Redes maiores com maior redundância se recuperam mais rapidamente. (4) A distribuição de betweenness centrality segue distribuição exponencial na maioria das redes. (5) Estudos anteriores focavam apenas em vulnerabilidade/robustez — este é o primeiro a abordar sistematicamente a fase de recuperação |
| Relação com o TCC | Conexão forte e direta. (1) Valida que métricas topológicas são bons preditores de resiliência — justifica nossa abordagem. (2) A metodologia de remoção progressiva de links é análoga à nossa remoção de nós. (3) O achado sobre betweenness seguir distribuição exponencial pode ser testado na nossa rede de SP. (4) O foco em recuperação complementa o foco em vulnerabilidade — são faces da mesma moeda |
| Citações úteis | Os autores demonstram que a capacidade de recuperação de redes de transporte público pode ser prevista por indicadores topológicos, com redes mais densas e eficientes apresentando maior resistência a perturbações |
| Limitações | (1) Analisa apenas redes de metrô, não ônibus — o TCC pode estender a abordagem pra ônibus. (2) Abordagem puramente topológica — não considera demanda de passageiros nem aspectos operacionais. (3) Não analisa equidade territorial |
| Usar no TCC como | Referência metodológica central — justifica a abordagem topológica e a simulação de resiliência. Citar na introdução e na metodologia |
