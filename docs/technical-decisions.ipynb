**Decisão:**

**Justificativa:**
- foco em vulnerabilidade e resiliência
  - para análise de vulnerabilidade, a topologia da rede (quais paradas/estação estão conectadas) é o que é relevante.
  - na remoção de nós, o nó desaparece, independente do número de ônibus/trem/metrô que passam por ele.
  - na resiliência estrutural da rede, o tempo de espera não é relevante.

**Consequências:**
- Métricas a serem consideradas:
  1. Tamanho da Maior Componente Conectada (LCC - *Largest Connected Component*): para Resiliência
    $$LCC_{ratio} = \frac{|LCC_{após\ remocao}|}{|Total\ de\ Nós|}$$
    - proporção de paradas que permanecem conectadas após ataques
    - como interpretar: $LCC = 0.8$ significa que 80% da rede ainda está conectada
    - não é afetada por pesos, pois é uma métrica topológica
  2. Centralidade de Intermediação (*Betweenness Centrality*): para Identificação de Nós Críticos
   $$$$

   - identifica paradas/estações críticas
   - remover nós em ordem decrescente de *betweenness* 

  3. Comprimento Médio dos Caminhos Mínimos (ASPL - *Average Shortest Path Length*): para Resiliência
   $$$$

  - número médio de paradas/estações entre qualquer par de nós
  - como interpretar: um aumento significativo no ASPL após ataques indica que a rede se tornou menos eficiente para os usuários, mesmo que ainda esteja conectada
  - não é afetada por pesos, pois é uma métrica topológica

  4. Diâmetro da Rede: para Resiliência
   $$$$
  - maior distância entre quaisquer dois nós na rede
