# Robustness Assessment of Public Transport Networks in Various Graph Representations: Systematic Review, Decision Support, and Case Study


| Campo | Conteúdo |
|:---|:---|
| Referência ABNT | PALK, J. et al. Robustness Assessment of Public Transport Networks in Various Graph Representations: Systematic Review, Decision Support, and Case Study. Networks, 2026. DOI: 10.1002/net.70021 |
| Autores | J. Palk et al. |
| Ano / Periódico | 2026 / Networks (Wiley) |
| Objetivo | Revisão sistemática das abordagens de representação em grafos para avaliação da robustez de redes de transporte público, fornecendo suporte à decisão sobre qual métrica e método usar |
| Metodologia | Revisão sistemática da literatura. Compara representações de grafos (L-space, P-space, C-space), métodos de avaliação (remoção sequencial de nós/arestas), e indicadores de robustez (tamanho do maior componente conexo, caminho médio mais curto, eficiência global) |
| Principais resultados | Ataques direcionados a nós com alta centralidade de intermediação ou de grau causam o impacto mais severo na estabilidade da rede, acelerando a fragmentação. Combinar múltiplas métricas (não depender só de uma) fornece avaliação mais robusta. Representação L-space (paradas como nós, conexões consecutivas como arestas) é a mais usada e adequada pra redes de ônibus |
| Relação com nosso TCC | Conexão DIRETA e forte. (1) Valida nossa escolha de representação L-space; (2) Confirma que betweenness é a métrica mais relevante pra vulnerabilidade; (3) Valida nossa metodologia de remoção progressiva de nós; (4) Sugere que devemos usar múltiplas métricas (o que já fazemos com degree + betweenness + closeness) |
| Citações úteis | A revisão confirma que a representação L-space é a mais adotada para redes de ônibus e que ataques direcionados baseados em centralidade de intermediação produzem o maior impacto na fragmentação da rede |
| Limitações | É uma revisão, não aplica a uma cidade específica. Não analisa equidade territorial |
| Usar no TCC como | Referência metodológica principal — justifica TODAS as escolhas metodológicas do nosso trabalho (representação do grafo, métricas, simulação). Citar na introdução e na metodologia |
