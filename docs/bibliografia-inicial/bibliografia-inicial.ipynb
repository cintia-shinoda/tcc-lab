Compara representações de grafos (L-space, P-space, C-space), métodos de avaliação (remoção sequencial de nós/arestas), e indicadores de robustez (tamanho do maior componente conexo, caminho médio mais curto, eficiência global) |
| 2 | [SONG, J.; DING, J.; GUI, X.; ZHU, Y. Indicadores calculados: degree, betweenness centrality, eficiência de rede e volume de fluxo de passageiros (PFV). Classificação dos nós em 6 clusters usando K-means. Simulação de ataque: remoção individual de cada nó e medição da redução na eficiência global da rede. Análise da distribuição geográfica dos nós críticos considerando a topografia montanhosa de Chongqing. Rede: 233 nós, 254 arestas, grau médio 2,197     ----Usa degree, betweenness + K-means pra classificar nós críticos. |
