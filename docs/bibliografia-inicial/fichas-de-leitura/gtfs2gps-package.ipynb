# Exploring the time geography of public transport networks with the gtfs2gps package 


| Campo | Conteúdo |
|:---|:---|
| Referência ABNT | PEREIRA, R. H. M.; ANDRADE, P. R.; BAZZO VIEIRA, J. P. Exploring the time geography of public transport networks with the gtfs2gps package. Journal of Geographical Systems, v. 25, p. 453-466, 2022. |
| Autores | Rafael H. M. Pereira (IPEA), Pedro R. Andrade (INPE), João Pedro Bazzo Vieira (IPEA) |
| Ano / Periódico | 2022 / Journal of Geographical Systems (Springer) |
| Objetivo | Apresentar o pacote gtfs2gps, uma ferramenta computacional que converte dados GTFS estáticos em tabelas de trajetórias espaço-temporais (formato GPS simulado), permitindo analisar padrões espaciais e temporais de redes de transporte público com resolução fina | 
| Base teórica | Time Geography (Hägerstrand, 1970) — framework conceitual que analisa como atividades humanas e padrões de mobilidade são restringidos pelo espaço e tempo. Conceitos-chave: space-time path, path bundling, space-time aquarium | 
| Metodologia | Pacote R open-source com computação paralela. Algoritmo em 6 etapas: (1) amostragem espacial de pontos ao longo das rotas, (2) snap de paradas às rotas, (3) cálculo de distância entre paradas, (4) cálculo de duração, (5) cálculo de velocidade, (6) interpolação espaço-temporal. Resolução espacial padrão: 100m. Caso de estudo: GTFS da EMTU São Paulo (jun/2019 a jan/2020) |
| Principais resultados | (1) Visualização 3D das trajetórias espaço-temporais de veículos em "aquário espaço-tempo". (2) Análise de desigualdade socioeconômica: áreas dos 20% mais ricos de SP têm frequência de serviço 28% maior que áreas de baixa renda no pico da manhã. (3) Pico de serviço em áreas de baixa renda começa ligeiramente mais cedo, possivelmente pra acomodar commutes mais longos da periferia | 
| Relação com o TCC | Conexão indireta mas relevante: (1) demonstra o potencial dos dados GTFS pra análise espaço-temporal — valida nosso uso de GTFS; (2) o achado sobre desigualdade de frequência entre faixas de renda reforça nossa P2 sobre equidade territorial; (3) usa dados de SP, mesmo contexto do nosso trabalho | 
| Citações úteis | O artigo mostra que a frequência de transporte público é 28% maior em áreas ricas vs pobres durante o pico da manhã — dado concreto pra justificar nossa análise de equidade | 
| Limitações(1) | Pacote em R, não Python; (2) GTFS estático não captura variabilidade real de tempo de viagem (congestionamento, acidentes); (3) Assume velocidade constante entre paradas quando GTFS só tem horário de partida e chegada; (4) Não faz análise de rede/grafos — foco é trajetória, não topologia | 
| Usar no TCC como | Referência bibliográfica na introdução — exemplo de ferramenta brasileira para análise de GTFS e evidência de desigualdade espaço-temporal no transporte de SP |
